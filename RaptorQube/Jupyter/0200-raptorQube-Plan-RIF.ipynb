# Design Basic RAPTOR RIF

```python
CoreSuperSteps=np.array([
    'Start',
    'Retrieve',
    'Assess',
    'Process',
    'Transform',
    'Organize',
    'Report',
    'End'])
```

```python
G=nx.DiGraph()
```

```python
MaxStep=CoreSuperSteps.shape[0]

for i in range(1, MaxStep):
    if i == 1:
        node1=CoreSuperSteps[i-1]
    else:
        node1=CoreSuperSteps[i-1] + ' Super Step'
    if i == MaxStep-1:
        node2=CoreSuperSteps[i]
    else:
        node2=CoreSuperSteps[i] + ' Super Step'
        
    G.add_node(
        node1,
        supersteptype=True
    )    
    
    G.add_node(
        node2,
        supersteptype=True
    )
    
    G.add_edge(
        node1, 
        node2, 
        weight=1000
    )
    
    if i < MaxStep-1:
        for j in range(8):
            node3 = '%s-%06d' % (CoreSuperSteps[i],j) 

            G.add_node(
                node3,
                crawlertype=True
            )

            G.add_edge(
                node1, 
                node3, 
                weight=1
            )

            G.add_edge(
                node3, 
                node2, 
                weight=1
            )
```
