### 300-Crawler-Input-Template-000001

```python
import datetime
nowStart = datetime.datetime.now()
```

```python
import pandas as pd
import os
```

```python
def InputData():
    dirname='../../100-DL/100-Raw-Zone/0100-External/100-Golden-Source/000005-Data-Source-F/'
    filename='First-Name-Boy-000001.csv'
    fullfilename=os.path.abspath(os.path.join(dirname, filename))
    print(fullfilename)
    dfIn=pd.read_csv(
        fullfilename,
        header=0,
        sep=',',
        encoding='latin1',
        skip_blank_lines=True,
        quotechar='"',
        verbose=False,
        low_memory=True
    )
    return dfIn
```

```python
InData = InputData()
```

```python
InData.shape
```

```python
InData
```

```python
def ProcessData(df0):
    df1 = df0.loc[df0.id.notnull()]
    return df1
```

```python
ProcData = ProcessData(InData)
```

```python
ProcData.shape
```

```python
ProcData
```

```python
def OutputData(df):
    dirname='../../100-DL/100-Raw-Zone/0200-Internal/000100-CSV/'
    filename='First-Name-Boy-000001.csv'
    fullfilename=os.path.abspath(os.path.join(dirname, filename))
    print(fullfilename)
    df.to_csv(
        fullfilename,
        header=0,
        index=False,
        sep=',',
        encoding='latin1',
        quotechar='"'
    )
```

```python
OutputData(ProcData)
```
