```python
import datetime
nowStart = datetime.datetime.now()
```

```python
import time
import gzip
import shutil
import os
```

```python
pathraptorQubename='./200-raptorQube/'
pathPackname='./900-RAPTOR-Pack/'
```

```python
if os.path.exists(pathraptorQubename):
    print('Removing existing raptorQube')
    shutil.rmtree(pathraptorQubename)
    time.sleep(5)
    
if not os.path.exists(pathraptorQubename):
    print('Creating new raptorQube')
    print(pathraptorQubename)
    os.makedirs(pathraptorQubename)
```

```python
# Unpack raptorQube
```

```python
filename='0100-raptorQube-RIF-DL-Setup.ipynb'
inFile=os.path.join(pathPackname, filename + '.gz')
outFile=dirname = os.path.join(pathraptorQubename, filename)
```

```python
import gzip
import shutil
with  gzip.open(inFile, 'rb') as f_in:
    with open(outFile, 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)
```
