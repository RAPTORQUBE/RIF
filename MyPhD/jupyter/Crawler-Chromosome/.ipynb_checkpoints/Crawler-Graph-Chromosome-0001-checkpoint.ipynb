## DNA Builder

```python
import networkx as nx
import sqlite3 as sq
from sqlite3 import Error
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dbnamedir='./000-RIF/200-Operational-Management-Layer/100-Crawler-Definitions/100-Crawler-Master-Service-Definitions/'
dbname='Crawler-Master-Service-Definitions.db'
dbfullname=os.path.join(dbnamedir, dbname)
print(dbfullname)
```

```python
conn = sq.connect(dbfullname)
```

```python
sSql="select * from DNAData"

sqlDf=pd.read_sql(sSql, conn)
print(sqlDf.shape)
print(sqlDf.columns)
```

```python
G = nx.DiGraph()
```

```python
edgesDf=sqlDf.copy(deep=True)
edgesDf.head()
```

```python
%%javascript
IPython.OutputArea.auto_scroll_threshold = 9999;
```

```python
for root, dirs, files in os.walk('./results', topdown=False):
    for name in files:
        print(os.path.join(root, name))
        filename, file_extension = os.path.splitext(name)
        if file_extension == '.jpg':
            os.remove(os.path.join(root, name))
```

```python
for i in range(edgesDf.shape[0]):
    print('%s%s' % ('\n', '='*100))
    print('Crawler set: %0d' % (i+1))
    print('='*100)
    
    rifFigName='./results/RAPTORQUBE-%06d.jpg' % (i+1)
    
    rif='RAPTORQUBE\n %06d' % (i+1)
    
    G = nx.DiGraph()
    
    G.add_edge(rif,edgesDf['CrawlerMasterYoke'][i])
    G.add_edge(edgesDf['CrawlerMasterYoke'][i],edgesDf['CrawlerSlaveYoke'][i])
    G.add_edge(edgesDf['CrawlerSlaveYoke'][i],edgesDf['CrawlerControlYoke'][i])
    
    
    G.add_edge(edgesDf['CrawlerControlYoke'][i],edgesDf['CrawlerInputChromosome'][i])
    G.add_edge(edgesDf['CrawlerControlYoke'][i],edgesDf['CrawlerProcessChromosome'][i])
    G.add_edge(edgesDf['CrawlerControlYoke'][i],edgesDf['CrawlerOutputChromosome'][i])
    
    
    G.add_edge(edgesDf['CrawlerInputChromosome'][i],edgesDf['CrawlerProcessChromosome'][i])
    G.add_edge(edgesDf['CrawlerProcessChromosome'][i],edgesDf['CrawlerOutputChromosome'][i])
    
    pos = nx.kamada_kawai_layout(G, scale=2)
    fig=plt.figure(figsize=(15,15))
    nx.draw_networkx(G, node_size=10000, node_color='lightgreen', font_size=12, pos=pos)
    limits = plt.axis('off')
    plt.show()
    
    fig.savefig(rifFigName, dpi=600)
```

```python
conn.close()
```
